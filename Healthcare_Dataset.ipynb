```python
import pandas as pd

df = pd.read_csv("/content/healthcare_dataset.csv")
df.head()
```

```python
df.dtypes
```

```python
 #Admission sütununu date formatına keçiririk-According to the date format of the receipt column
df['Date of Admission']=pd.to_datetime(df['Date of Admission'])
```

```python
df['Discharge Date']=pd.to_datetime(df['Discharge Date'])
```

```python
df['Medical Condition'].value_counts().reset_index()
```

```python
#Sətir və sütun sayı -- Number of Rows and colunms
print('Colunms',df.shape[1])
print('Rows',df.shape[0])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(x='Medical Condition',y='Billing Amount', data=df)
plt.title('Medical Condition üçün boxplot')
plt.show()
```

```python
sns.boxplot(x='Billing Amount',data=df)
plt.show()
```

```python
#Billing Amount üçün outliers -- Outliers for Billing Amount

Q1=df['Billing Amount'].quantile(0.25)
Q3=df['Billing Amount'].quantile(0.75)

IQR=Q3-Q1

outliers=df[(df['Billing Amount']<Q1-1.5*IQR) | (df['Billing Amount']>Q3-1.5*IQR)]
outliers
```

```python
#Xəstəliklərin sayı-Number of diseases
df['Medical Condition'].hist(bins=12)
plt.title("Medical Condition")
plt.xlabel='Medical Condition'
plt.ylabel='Count'
plt.show()
```

```python
#Admission Type növləri
sns.countplot(x='Admission Type',data=df)
plt.show()
```

```python
#Test results-a görə Pie Chart -- Pie Chart according to test results
df['Test Results'].value_counts().plot.pie(autopct='%1.1f%%',startangle=90)
plt.title('Pie chart for test results')
```

```python
#Numeric dəyərlər üçün correlation matrix
corr=df.corr(numeric_only=True)
sns.heatmap(corr,annot=True)
plt.show()
```

```python
df['Admission Year']=df['Date of Admission'].dt.year
sns.countplot(x='Admission Year',data=df)
plt.title("Hər il üçün ümumi say-Total number for each year")
```

```python
#Hər xəstəliyə görə test nəticələri-Test results for each disease
sns.catplot(x='Medical Condition',hue='Test Results',kind='count',aspect=2,data=df)
plt.title("Hər xəstəliyə görə test nəticələri-Test results for each disease")
plt.show()
```

```python
#Hər cins üzrə Billing Amount -- Billing Amount per gender
df.groupby('Gender')['Billing Amount'].mean().reset_index()
```

```python
#Hər medical condition üçün Billing Amount-un mean(average),max,min və median qiymətləri
billing_status=df.groupby('Medical Condition')['Billing Amount'].agg(['mean','median','max','min']).reset_index()
billing_status
```

```python
#Xəstələrin qalma müddəti adlı sütun yaradırıq--Creating a column called Patient Length of Stay
df['qalma muddeti']=(df['Discharge Date']-df['Date of Admission']).dt.days
df['qalma muddeti']
```

```python
df.groupby(['Medical Condition','Admission Type'])['qalma muddeti'].mean().reset_index()
```

```python
#Yaşı 30-dan çox olan pasiyentlər -- Patients over 30 years of age
df[df['Age']>=30][['Name']]
```

```python
#Hər yaş aralığı və xəstəliyə görə pasiyentlərin sayı  -- Number of patients by age group and disease
bins=[0,18,35,50,65,100]
labels=['0-18','19-35','36-50','51-65','66+']
df['yas_qrupu']=pd.cut(df['Age'],bins=bins,labels=labels)

age_condition=df.groupby(['yas_qrupu','Medical Condition']).size().unstack(fill_value=0)
age_condition
```
